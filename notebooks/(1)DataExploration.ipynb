```python
import pandas as pd

# 1. Load your dataset
df = pd.read_csv("../data/base/2023_brazil_stocks.csv")

# (Optional) Check the first few rows to confirm column names
print(df.head())

# 2. Choose the top 5 companies
selected_tickers = ['PETR4', 'VALE3', 'ITUB4', 'BBDC4', 'ABEV3']  # example tickers

# 3. Filter the DataFrame
filtered_df = df[df['ticker'].isin(selected_tickers)]

# 4. Save the filtered dataset to a new CSV file
filtered_df.to_csv("data/filtered_stocks_top_5.csv", index=False)

print(f"Filtered dataset saved: {len(filtered_df)} rows for {len(selected_tickers)} companies.")

df['ticker'].value_counts()

```

```python
import pandas as pd

# Load your filtered stock data
df = pd.read_csv("../data/filtered_stocks_top_5.csv")

# Load full dataset (for the IBOV rows)
full_df = pd.read_csv("../data/base/2023_brazil_stocks.csv")

# Extract only the IBOVESPA index rows
ibov_df = full_df[full_df['ticker'].isin(['IBOV', 'IBOVE', 'IBOVE11', 'IBOVF11'])]

# Keep only necessary columns (e.g., date and close price)
ibov_df = ibov_df[['date', 'close']].rename(columns={'close': 'ibov_close'})

# Merge index data with your 5 companies by date
merged_df = df.merge(ibov_df, on='date', how='left')

# Save the result
merged_df.to_csv("data/brazil_stocks_top_5_with_ibov.csv", index=False)

print(f"Merged dataset with IBOVESPA index data. Total rows: {len(merged_df)}")

merged_df.head()
```
